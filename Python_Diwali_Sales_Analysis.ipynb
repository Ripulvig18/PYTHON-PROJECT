# EXPLORATORY DATA  ANALYSIS

```python
ax = sns.countplot(x = "Gender",data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gender)

for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs, we can conclude that most number of buyers are females and purchasing power of females are greater than males.

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
ax = sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

From above graphs, we can see that most of the buyers are of age group between 26-35 yrs female.

```python
state_sales = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x ='State', y='Orders', data = state_sales)
```

```python
state_amount = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x ='State', y='Amount', data = state_amount)
```

From above graphs, we can conclude that mostof the orders are from Uttar Pradesh, Maharashtra and Karnataka reapectively.  

```python
ax = sns.countplot(data = df , x = 'Marital_Status' )

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age_gen = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_age_gen, x = 'Marital_Status', y='Amount', hue = 'Gender')
```

From above graph, we can conclude that most of the buyers are married (women) and they have high purchasing power too.

```python
ax = sns.countplot(data = df, x = 'Occupation')

sns.set(rc = {'figure.figsize':(30,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occup = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_occup, x = 'Occupation', y='Amount')
```

From above graphs, we can conclude that most of the buyers are IT, Healthcare and Aviation sector.

```python
ax = sns.countplot(data = df , x = 'Product_Category' )

sns.set(rc={'figure.figsize':(28,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occup = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(28,5)})
sns.barplot(data = sales_occup, x = 'Product_Category', y='Amount')
```

From above graphs, we can conclude that most of the sold products are food, clothing and Electronics category.

```python
productid_sales = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(28,5)})
sns.barplot(data = productid_sales, x = 'Product_ID', y='Orders')
```

```python
fig, ax = plt.subplots (figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind = "bar")
```

Married women age group 26-35 yrs from UP,Maharashtra an Karnataka working in IT, Healthcare an Aviation are more likely to buy products from Foo, Clothing and Electronics category.
